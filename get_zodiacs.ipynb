# Imports

```python
import csv
from kerykeion import Report, AstrologicalSubject
import pandas as pd
```

# Store in new csv 

```python
input_players = 'football_bdays.csv'
output_players = 'football_zodiacs.csv'
```

# Get the players sun, moon, and rising signs and add to new CSV

```python
with open(input_players, mode='r', newline='', encoding='utf-8') as infile, \
     open(output_players, mode='w', newline='', encoding='utf-8') as outfile:
    
    reader = csv.DictReader(infile)
    
    # Add new cols for signs
    fieldnames = reader.fieldnames + ['sun_sign', 'moon_sign', 'rising_sign']
    
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for row in reader:
        # get values from for each player
        name = row['name']
        year = int(row['year'])
        month = int(row['month'])
        day = int(row['day'])
        
        subject = AstrologicalSubject(name, year, month, day)
        
        report = Report(subject)
        
        # get sign values for each attribute
        sun_sign = subject.sun.sign
        moon_sign = subject.moon.sign
        rising_sign = subject.first_house.sign

        print(f"Report for {name}:")
        print(f"  Sun: {sun_sign}")
        print(f"  Moon: {moon_sign}")
        print(f"  Rising: {rising_sign}")
        print("-" * 40)
        
        # Add the new values to the row
        row['sun_sign'] = sun_sign
        row['moon_sign'] = moon_sign
        row['rising_sign'] = rising_sign
        
        # Write the updated row to the output file
        writer.writerow(row)

print(f"Finished adding zodiac signs to file: '{output_players}'")
```

# Remove birthdays from csv

```python

df = pd.read_csv('football_zodiacs.csv')

# drop bday cols
df = df.drop(columns=['year', 'month', 'day'])

df.to_csv('football_zodiacs.csv', index=False)
```
